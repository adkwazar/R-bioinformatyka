<h4> Higher dimensional spaces

```python
import numpy as np
import matplotlib.pyplot as plt
```

- 2D Gauss

```python
means = np.zeros(2)
sigma_cov = np.identity(2)
N = 100
data = np.random.multivariate_normal(means, sigma_cov, 100) 
distances = []
for i in range(N):
    for j in range(i,N):
        distances.append(np.linalg.norm(data[i]-data[j]))
        
plt.hist(distances)
plt.show()
```

<h4> Eigenvalues and eigenvectors

```python
from numpy import linalg as LA

A = np.array([[1,2],
              [5,6]])
lam, v = LA.eig(A) 

print(lam) #eigenvalues
print()
print(v) #eigenvectors (columns)
```

```python
print(np.dot(A,v[:,0]))
print()
print(lam[0]*v[:,0])
```

<h4> PageRank alogorithm

```python
import numpy as np
from numpy import linalg as LA

A = np.array([[0,0,0,0,0.5,0],[0.25,0,0,0,0,0],[0,0.5,0,0,0,0],[0.25,0.5,0,0,0.5,0],[0.25,0,1,1,0,1],[0.25,0,0,0,0,0]])
lam, v = LA.eig(A) 

print(lam)

import matplotlib.pyplot as plt

#polar coordinates

def polar(z):
    a= z.real #x coordinate
    b= z.imag #y coordinate
    r = np.sqrt(a**2+b**2) #radius
    theta = np.arctan2(b,a) #angle
    return r,theta 


for x in lam:
    plt.polar([0,polar(x)[1]],[0,polar(x)[0]],marker='o')

print()
print(np.abs(lam)) #we take the highest eigenvalue
print()
print(np.abs(v[:,0])) #... and its corresponding eigenvector
```

<h4> Distance based methods

```python
from scipy.optimize import minimize

fun = lambda w: abs((w[0]-w[2])**2+(w[1]-w[3])**2 - 8)+\
                abs((w[0]-w[4])**2+(w[1]-w[5])**2 - 8)+\
                abs((w[0]-w[6])**2+(w[1]-w[7])**2 - 8)+\
                abs((w[2]-w[4])**2+(w[3]-w[5])**2 - 8)+\
                abs((w[2]-w[6])**2+(w[3]-w[7])**2 - 8)+\
                abs((w[4]-w[6])**2+(w[5]-w[7])**2 - 8)

res = minimize(fun,x0=(0,0,2*2**0.5,0,2**0.5,2,2**0.5,-2), method='SLSQP')
print(fun(res.x))
print(res.x) #where is the minimum?
```

```python
import matplotlib.pyplot as plt
plt.scatter(res.x[::2],res.x[1::2])
plt.show()
```

```python
#other solution

from scipy.optimize import minimize

fun = lambda w: abs(w[0]**2 - 8)+\
                4*abs(0.25*w[0]**2+w[1]**2- 8)+\
                abs(4*w[1]**2 - 8)

res = minimize(fun,x0=(1.4,2.7), method='SLSQP')
print(res.x)
print(fun(res.x))
```

<h4> FDA

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([-2,1,2,3])
y1 = 3*x+5
y2 = 3*x-5

plt.scatter(x,y1)
plt.scatter(x,y2)
plt.show()
```

```python
x1 = np.c_[x, y1]
x2 = np.c_[x, y2]
print(x1)
print()
print(x2)
```

```python
u1 = np.mean(x1, axis = 0)
u2 = np.mean(x2, axis = 0)

s1 = np.cov(x1.T)
s2 = np.cov(x2.T)
SW= s1+s2

z = u1-u2
SB = np.outer(z,z.T)

def J(v):
    n = np.dot(np.dot(v.T, SB), v)
    d = np.dot(np.dot(v.T, SW), v)
    return n/d
```

```python
#to have a vector with corodinates between -1 and 1
cons = ({'type': 'ineq', 'fun': lambda w: 1-w[0]},
        {'type': 'ineq', 'fun': lambda w: 1-w[1]},
        {'type': 'ineq', 'fun': lambda w: w[0]+1},
        {'type': 'ineq', 'fun': lambda w: w[1]+1})

fun2 = lambda w: -J(w)
res = minimize(fun2, (0.5,0.2), method='SLSQP',constraints=cons)
print(res.x) #where is the minimum?
```
